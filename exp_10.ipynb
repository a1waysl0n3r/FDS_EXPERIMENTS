```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\latai\Downloads\tips.csv')
data
```

```python
x = data['total_bill']
y = data['tip']
sns.scatterplot(data = data, x = 'total_bill', y = 'tip')
plt.title("Total Bill vs Tip amount")
plt.show()
```

```python
x = data['total_bill']
y = data['tip']
sns.scatterplot(data = data, x = 'total_bill', y = 'tip', hue = 'day', style = 'smoker')
plt.title("Total Bill vs Tip amount")
plt.show()
```

```python
x = data['total_bill']
y = data['tip']
plt.figure(figsize=(12, 6))
sns.lineplot(data = data, x = 'total_bill', y = 'tip', hue = 'time')
plt.title("Total Bill vs Tip amount")
plt.show()
```

```python
# Customizing line plot
sns.lineplot(data=data, x='size', y='total_bill', hue='day', style='sex')
plt.title('Line plot with Different Categories')
plt.show()
```

```python
sns.kdeplot(data = data, x = 'total_bill', fill = True)
plt.title("Density distribution of Bill")
```

```python
sns.kdeplot(data = data, x = 'total_bill', fill = True)
sns.kdeplot(data = data, x = 'tip', fill = True)
plt.show()
```
