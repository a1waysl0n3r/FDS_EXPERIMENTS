```python
data1 = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'DepartmentID': [1, 2, 3, 1, 3],
    'Name': ['John Doe', 'Jane Smith', 'Bob Johnson', 'Alice Brown', 'Mike Davis'],
    'Salary': [70000, 60000, 45000, 80000, 50000]
}
data2 = {
    'EmployeeID': [101, 102, 103, 106],
    'DepartmentID': [1, 2, 3, 4],
    'Project': ['Alpha', 'Beta', 'Gamma', 'Delta'],
    'HoursWorked': [35, 40, 25, 30]
}
dataa = pd.DataFrame(data1)
datab = pd.DataFrame(data2)
```

```python
print(dataa)
print(datab)
```

```python
data = pd.merge(dataa, datab, on = 'EmployeeID', how = 'left')
data
```

```python
datax = pd.merge(dataa, datab, on = 'EmployeeID', how = 'right')
datax
```

```python
datay = pd.merge(dataa, datab, on = 'EmployeeID', how = 'inner')
datay
```

```python
dataz = pd.merge(dataa, datab, on = 'EmployeeID', how = 'outer')
dataz
```
