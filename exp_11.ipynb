```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\latai\Downloads\tips.csv')
data
```

```python
sns.barplot(data=data, x='day', y='total_bill', hue='smoker')
plt.title('Average Total Bill per Day by Smoking Habits')
plt.show()
```

```python
sns.histplot(data=data, x='total_bill', bins=20)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.histplot(data=data, x='tip', bins=20)
plt.title('Distribution of Tip')
plt.show()
```

```python
sns.kdeplot(data = data, x = 'total_bill', fill = True)
plt.title("Density distribution of Bill")
```

```python
sns.kdeplot(data = data, x = 'total_bill', fill = True)
sns.kdeplot(data = data, x = 'tip', fill = True)
plt.show()
```
